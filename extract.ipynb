```python
import json

with open("notchian/generated/reports/blocks.json", "r") as f:
    data = json.load(f)
```

```python
oak_stairs = data["minecraft:oak_stairs"]
states = oak_stairs['states']
```

```python
for s in states:
    properties = s['properties']
    if properties['half'] == 'bottom' and properties['shape'] == "straight" and properties['waterlogged'] == 'false':
        print(s['id'], properties['facing'])
```
